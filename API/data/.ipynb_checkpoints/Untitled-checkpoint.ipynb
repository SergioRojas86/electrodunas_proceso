```python
anomalias = df[df.is_outlier_lof==True].groupby('Sector_Economico').agg(cantidad=('is_outlier_lof','count')).reset_index()
```

```python

anomalías = df[df['is_outlier_lof']]
conteo_por_sector = Counter(anomalías['Sector_Economico'])
total_anomalías = len(anomalías)
# Calcular porcentajes
data = [{'Sector': sector, 'Porcentaje': int(round((count / total_anomalías) * 100,0))} for sector, count in conteo_por_sector.items()]
```
